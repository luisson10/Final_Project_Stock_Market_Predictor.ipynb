## Load The Stock Market Data

```python
#List of companies using their ticker symbol
company1 = 'INTC.MX' #First company Intel
company2 = 'AAPL.MX' #Second company Apple
company3 = 'DIS.MX' #Third company Disney
company4 = 'LABB.MX' #Fourt company Genoma LAB

companies_list = ['INTC.MX', 'AAPL.MX', 'DIS.MX', 'LABB.MX']

for company in companies_list: #for loop in oder to get output from all the companies 
  

  start = dt.datetime(2015,1,1) #Get data from 1st January of 2015
  end = dt.datetime(2022,1,1) #Stop getting data at the 1st January of 2020
  #end = dt.datetime.now()

  #data = pdr.DataReader(company, 'yahoo', start, end) #To look for the data online on yahoo
  data = yf.download(company, start, end)
  data #To show the data

  #To plot the stock market data obtained
  plt.figure(figsize=(16,8)) #To make the plot bigger
  plt.plot(data['Close'], color="Blue", label=f"Actual {company} Price") #Plots the actual stock
  plt.title(f"{company} Close Price", fontsize=15) #Title
  plt.xlabel('Time', fontsize=15) #X-Label
  plt.ylabel(f"{company} Close Price in USD", fontsize=15) #Y-Label
  plt.legend()
  plt.grid()
  plt.show()

  #To get number of rows to train the model on
  data = data.filter(['Close']) #To only get the Close Values of the stock from data
  dataset = data.values #To convert dataframe to numpy array

  training_data_len = math.ceil(len(dataset)* .8) 

  #Scale the Data Values
  scaler = MinMaxScaler(feature_range=(0,1))
  #scaled_data = scaler.fit_transform(data['Close'].values.reshape(-1,1))
  scaled_data = scaler.fit_transform(dataset)

  train_data = scaled_data[0:training_data_len , :]

  prediction_days = 60 
  #Using 60 days in order to predict 1 day in the future
  #If we decrease the number of days, our graph will be more sensitive on short term fluctuations
  #If we increase the number of days, our graph will be more sensitive on long term fluctuations

  x_train = [] #Variable to train the data
  y_train = [] #Variable to train the data
  for x in range(prediction_days, len(scaled_data)):
    x_train.append(scaled_data[x-prediction_days:x, 0])
    y_train.append(scaled_data[x, 0])

  x_train, y_train = np.array(x_train), np.array(y_train) #To convert into numpy arrays
  x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1)) #To reshape the data

  x_train1 = [] #Variable to train the data
  y_train1 = [] #Variable to train the data
  for x in range(prediction_days, len(train_data)):
    x_train1.append(train_data[x-prediction_days:x, 0])
    y_train1.append(train_data[x, 0])

  x_train1, y_train1 = np.array(x_train1), np.array(y_train1) #To convert into numpy arrays
  x_train1 = np.reshape(x_train1, (x_train1.shape[0], x_train1.shape[1], 1)) #To reshape the data

  #Build the Model
  model = Sequential() #This is a Recurrent neural networks (RNN) Model Using Keras
  #Sequential Learning
  #To iterate over the timesteps of a sequence, while maintaining an 
  #internal state that encodes information about the timesteps it has seen so far.

  #This is a Long Short-Term Memory (LSTM) model using Keras
  model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1], 1)))
  #we add a LSTM layer with 50 neurons and since first layer we give an input shape
  model.add(Dropout(0.2))
  model.add(LSTM(units=50, return_sequences=True))
  #Second LSTM layer with also 50 neurons
  model.add(Dropout(0.2))
  model.add(LSTM(units=50))
  #Third LSTM layer with also 50 neurons
  model.add(Dropout(0.2))

  model.add(Dense(units=25))
  #Dense layer with 25 neurons
  model.add(Dense(units=1)) #prediction of the next closing value

  #To compile the Model
  model.compile(optimizer='adam', loss='mean_squared_error')
  #Optimizer is to learn and do better from the loss function
  #Loss function is to see how well the model performs in the training

  #To train the Model
  model.fit(x_train, y_train, epochs=10, batch_size=32)

  #Test model accuracy
  # Load test data
  test_start = dt.datetime(2020,1,1) #Tells us when to start looking for the test data online (2020)
  test_end = dt.datetime.now() #Tells us when to stop looking at data online (now)

  #test_data = pdr.DataReader(company, 'yahoo', test_start, test_end) #To look for the data online on yahoo
  test_data = yf.download(company, test_start, test_end)
  actual_prices = test_data['Close'].values

  total_dataset = pd.concat((data['Close'], test_data['Close']), axis=0)

  model_inputs = total_dataset[len(total_dataset)-len(test_data)- prediction_days:].values
  model_inputs = model_inputs.reshape(-1, 1)
  model_inputs = scaler.transform(model_inputs) #To transform the data to scalar form

  #Make prediction on test data
  test_data = scaled_data[training_data_len - prediction_days: , :]
  x_test = [] #New valiable
  y_test = dataset[training_data_len:, :]

  for x in range(prediction_days, len(test_data)):
    x_test.append(test_data[x-prediction_days:x, 0])
  #Now our y_test contains the past 60 values from the actual data set (Not Scaled)

  x_test = np.array(x_test) #Convert the data into a numpy array
  x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1)) #Reshape the data

  x_test1 = [] #New valiable

  for x in range(prediction_days, len(model_inputs)):
    x_test1.append(model_inputs[x-prediction_days:x, 0])

  x_test1 = np.array(x_test1) #Convert the data into a numpy array
  x_test1 = np.reshape(x_test1, (x_test1.shape[0], x_test1.shape[1], 1)) #Reshape the data

  #To get the models predicted price values
  predicted_prices = model.predict(x_test)
  predicted_prices = scaler.inverse_transform(predicted_prices) #To inverse transform the data (unscaling the values)

  predicted_prices1 = model.predict(x_test1)
  predicted_prices1 = scaler.inverse_transform(predicted_prices1)

  #To plot the test predictions
  plt.figure(figsize=(16,8)) #To make the plot bigger
  plt.plot(actual_prices, color="black", label=f"Actual {company} Price") #Plots the actual stock
  plt.plot(predicted_prices1, color="green", label=f"Predicted {company} Price") #Plots the our predicted stock
  plt.title(f"{company} Share Price", fontsize=15) #Title
  plt.xlabel('Time', fontsize=15) #X-Label
  plt.ylabel(f"{company} Share Price in USD", fontsize=15) #Y-Label
  plt.legend() #Legend
  plt.grid() #Grid
  plt.show()

  train= data[:training_data_len]
  valid= data[training_data_len:]
  valid['Predictions'] = predicted_prices # another colum

  plt.figure(figsize=(16,8)) #To make the plot bigger
  plt.plot(train['Close'], color="blue")
  plt.plot(valid['Close'], color="black")
  plt.plot(valid['Predictions'], color="red")
  plt.title(f"{company} Close Price", fontsize=15) #Title
  plt.xlabel('Year', fontsize=15) #X-Label
  plt.ylabel(f"{company} Close Price in USD", fontsize=15) #Y-Label
  plt.legend(['Train Data','Actual','Predictions']) #Legend
  plt.grid() #Grid
  plt.show()

  #To show the Actual and Predicted Values
  valid

  #To predict the next day
  real_data = [model_inputs[len(model_inputs)+1-prediction_days:len(model_inputs+1), 0]] 
  real_data = np.array(real_data)
  real_data=np.reshape(real_data, (real_data.shape[0], real_data.shape[1],1)) #To reshape the real data fromat

  prediction=model.predict(real_data) #Using real data as the input
  prediction = scaler.inverse_transform(prediction) 
  print(f"The Predicted Stock price for tomorrow will be: {prediction}")
```
